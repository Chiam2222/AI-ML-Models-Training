## Reshape And Transpose a Single Image

Our image is a single dimension array of size 3072. First 1024 entries of the array are of Red channel, next 1024 entries are of Green channel, and last 1024 entries are of Blue channel. Total 3072 entries are of three RGB channels. 

First we reshape the image/ array into (3,32,32) using image.reshape(3,32,32). (32,32) for 32x32 =1024 entries and 3 for RGB channels.

```python
image = data_batch_1['data'][0]
image = image.reshape(3,32,32)
print(image.shape)
```

Next we transpose the reshaped image to get the image of shape (32,32,3). We use image.transpose(1,2,0).

```python
image = image.transpose(1,2,0)
print(image.shape)
```

"Conver L" meaning: This image mode has one channel that can take any value between 0 and 255 representing white, black and all the shades of gray in between. It’s an 8-bit grayscale image mode. L stands for Luminance channel.

L mode only uses one value between 0-255 for each pixel (8-bit).


ITU-R 601 7th Edition Construction of Luminance formula:


L = R * 299/1000 + G * 587/1000 + B * 114/1000

By iterating through each pixel you can convert 24-bit to 8-bit or 3 channel to 1 channel for each pixel by using the formula above.

Change the colour images into grey images and visualise the first 50 of the training images.

```python
import matplotlib.pyplot as plt
import numpy as np

images = data_batch_1['data']
images = images.reshape(len(images),3,32,32).transpose(0,2,3,1)
labels = data_batch_1['labels']
label_names = meta_data['label_names']

rows, columns = 5, 10
imageId = np.random.randint(0, len(images), rows * columns)
images = images[imageId]
labels = [labels[i] for i in imageId]

fig=plt.figure(figsize=(20, 10))

for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow((Image.fromarray(images[i-1]).convert("L")), cmap = mpl.cm.binary, interpolation="nearest")
    plt.xticks([])
    plt.yticks([])
    plt.title("{}"
          .format(label_names[labels[i-1]]))
plt.show()
```

Verify the first 50 images black and white values are identical to 1.

```python
check_identical = []
count0 = 0
count1 = 0
for i in range(50):
    col_img = np.array(Image.fromarray(data_batch_1['data'][i].reshape(-1,32,32).T, mode='RGB').convert("L"))
    grey_img = np.array(batch1.iloc[i, :-1].values.reshape(-1, 32, 32)) 
    isZero = col_img - grey_img
    checkIsZero = not np.any(isZero)
    if checkIsZero == 0:
        check_identical.append(0)
        count0 += 1
    else:
        check_identical.append(1)
        count1 += 0
    i+=1
    
print(check_identical)
print("There are", count0, "zeros")
print("There are", count1, "ones")

```
